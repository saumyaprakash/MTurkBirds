# Bird Stats
Python Script for getting stats on the bird database that we have

```python
chosen_attributes = ['309', '310', '311', '312']
chosen_attribute_strings = {"309" : "Solid", "310" : "Spotted", "311" : "Striped", "312" : "Multi-colored"}
attribute_strings = ["Solid", "Spotted", "Striped", "Multi-colored"]    
```

```python
#create a bird species info dictionary
species_dict = {}
with open('./CUB_200_2011/CUB_200_2011/classes.txt') as f:
    for line in f:   
        info = line.replace('\n', '').split(" ") #class id, class name
        species_dict[info[0]] = {}
        species_dict[info[0]]["name"] = info[1]
        species_dict[info[0]]["images"] = []
        for attrib in chosen_attributes:
            species_dict[info[0]][attrib+"_count"] = 0
        
# species info needs name, list of ids, later add the percent per label and confidence per label
print(species_dict['1'])
```

```python
image_dict = {} #map id to file name, class, attributes
with open('./CUB_200_2011/CUB_200_2011/images.txt') as f:
    for line in f: 
        info = line.replace('\n', '').split(" ") #image id,  file name
        image_dict[info[0]] = {}
        image_dict[info[0]]["file_path"] = info[1]
print(image_dict['1'])
```

```python
with open('./CUB_200_2011/CUB_200_2011/image_class_labels.txt') as f:
    for line in f:
        info = line.replace('\n', '').split(" ") #image id, class id
        image_dict[info[0]]["class_id"] = info[1]
        species_dict[info[1]]["images"].append(info[0])
print(species_dict['1'])
```

```python
#add the information to the appropriate places
time_spent = []
with open('./CUB_200_2011/CUB_200_2011/attributes/image_attribute_labels.txt') as f:
    for line in f: 
        info = line.replace('\n', '').split(" ") #image id, attribute id, is present, certainty id, time
        species = image_dict[info[0]]["class_id"]
        if (info[1] in chosen_attributes) and (info[2] == '1'):
            #target attribute present for image
            image_dict[info[0]]['wing_pattern'] = chosen_attribute_strings[info[1]]
            image_dict[info[0]]['wing_pattern_confidence'] = info[3]
            species_dict[species][(info[1]+"_count")] += 1
            time_spent.append(info[4])
        
# print(image_dict['3'])
# print(species_dict['1'])
# print(time_spent)
```

```python
for key in species_dict.keys():
    curr_name = species_dict[key]['name']
    solid = species_dict[key]['309_count']
    spotted = species_dict[key]['310_count']
    striped = species_dict[key]['311_count']
    multi = species_dict[key]['312_count']
    total = solid + spotted + striped + multi
    feat_counts = [solid, spotted, striped, multi]
    majority = attribute_strings[feat_counts.index(max(feat_counts))]
    species_dict[key]['PrimaryPattern'] = majority
    species_dict[key]['TotalLabeled'] = total
    species_dict[key]['TotalImages'] = len(species_dict[key]['images'])
    #get a list of birds that fit in the majority
    species_dict[key]['images_in_majority'] = []
    for image in species_dict[key]['images']:
        if image_dict[image].get('wing_pattern') == majority:
            species_dict[key]['images_in_majority'].append(image)
    print('{}: {}% labeled, Majority {} | Solid {}% | Spotted {}% | Striped {}% | Multi {}%'.format( 
          curr_name, round(total/len(species_dict[key]['images']) * 100, 2), majority, round(solid / total * 100, 2),
        round(spotted / total * 100,2), round(striped / total * 100, 2), round(multi / total * 100, 2)))
    
```
